```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Optimal prefix/suffix ratio

```python
prefix_sizes = [30, 40, 50, 60, 70, 80, 90]
correct_completions = [30, 34, 36, 36, 34, 33, 28]
valid_completions = [58, 60, 63, 61, 62, 61, 55]

prefix_suffix_ratios = [f"{prefix_size}/{100-prefix_size}" for prefix_size in prefix_sizes]

fig, ax1 = plt.subplots(dpi=150)

# Plot for Correct Completions
ax1.plot(prefix_sizes, correct_completions, color="blue", label="Correct Completions")
ax1.set_xlabel("Prefix/Suffix Ratio")
ax1.set_ylabel("Correct Completions, %")

# Set the x-ticks and labels
ax1.set_xticks(prefix_sizes)
ax1.set_xticklabels(prefix_suffix_ratios)

# Create a twin axis for Valid Completions
ax2 = ax1.twinx()
ax2.plot(prefix_sizes, valid_completions, color="green", label="Valid Completions")
ax2.set_ylabel("Valid Completions, %")

# Adjust the y-axis scale
ax1.set_ylim([min(correct_completions) - 2, max(correct_completions) + 2])
ax2.set_ylim([min(valid_completions) - 2, max(valid_completions) + 2])

# Combine the legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper right")

plt.show()
```

### Optimal context size


```python
context_sizes = [1024, 2048, 3072, 4096, 5120, 6144, 7168]
correct_completions = [12, 30, 34, 36, 36, 35, 35]
valid_completions = [51, 62, 61, 63, 61, 63, 64]
generation_speeds = [10.95, 12.76, 11.6, 12.52, 13.01, 13.8, 14.73]

# Create a dataframe with the data
df = pd.DataFrame({'Context Sizes': context_sizes,
                   'Correct Completions': correct_completions,
                   'Valid Completions': valid_completions,
                   'Generation Speeds': [1000/speed for speed in generation_speeds]})

# Plot the data
fig, ax1 = plt.subplots(dpi=150)

# Plot for Correct and Valid Completions
ax1.plot(df['Context Sizes'], df['Correct Completions'], color="blue", label="Correct Completions")
ax1.plot(df['Context Sizes'], df['Valid Completions'], color="green", label="Valid Completions")
ax1.set_xlabel("Context Size, tokens")
ax1.set_ylabel("Completions, %")

# Set the x-ticks and labels
ax1.set_xticks(context_sizes)
ax1.set_xticklabels(context_sizes)

# Adjust the y-axis scale
# ax1.set_ylim([min(correct_completions) - 20, max(valid_completions) + 20])

# Create a twin axis for Generation Speeds
ax2 = ax1.twinx()
ax2.plot(df['Context Sizes'], df['Generation Speeds'], color="orange", label="Generation Speed")
ax2.set_ylabel("Generation Speed, char/s")

# Combine the legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper right")

plt.show()
```

### Metadata in components

```python
context_sizes = [4096, 5120, 6144, 7168]
correct_completions = [27, 38, 39, 35]
valid_completions = [59, 66, 64, 60]
correct_completions_without_components = [36, 36, 35, 35]
valid_completions_without_components = [63, 61, 63, 64]

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), dpi=150)

# Plot for correct completions with and without components
ax1.plot(context_sizes, correct_completions, color="blue", label="With Components")
ax1.plot(context_sizes, correct_completions_without_components, color="blue", linestyle="dotted", label="Without Components")
ax1.set_xlabel("Context Size, tokens")
ax1.set_ylabel("Correct Completions, %")
ax1.set_title("Correct Completions")
ax1.legend()

# Set the x-ticks and labels
ax1.set_xticks(context_sizes)
ax1.set_xticklabels(context_sizes)

# Plot for valid completions with and without components
ax2.plot(context_sizes, valid_completions, color="green", label="With Components")
ax2.plot(context_sizes, valid_completions_without_components, color="green", linestyle="dotted", label="Without Components")
ax2.set_xlabel("Context Size, tokens")
ax2.set_ylabel("Valid Completions, %")
ax2.set_title("Valid Completions")
ax2.legend()

# Set the x-ticks and labels
ax2.set_xticks(context_sizes)
ax2.set_xticklabels(context_sizes)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

### PSM vs SPM

### Autoregressive mode

```python
context_sizes = [1024, 2048, 3072, 4096, 5120, 6144, 7168]
correct_completions_ar = [1, 7, 10, 10, 14, 14, 15]
valid_completions_ar = [13, 19, 18, 24, 25, 25, 29]
correct_completions = [12, 30, 34, 36, 36, 35, 35]
valid_completions = [51, 62, 61, 63, 61, 63, 64]

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), dpi=150)

# Plot for correct completions with and without components
ax1.plot(context_sizes, correct_completions_ar, color="blue", label="AR")
ax1.plot(context_sizes, correct_completions, color="blue", linestyle="dotted", label="Infilling")
ax1.set_xlabel("Context Size, tokens")
ax1.set_ylabel("Correct Completions, %")
ax1.set_title("Correct Completions")
ax1.legend()

# Set the x-ticks and labels
ax1.set_xticks(context_sizes)
ax1.set_xticklabels(context_sizes)

# Plot for valid completions with and without components
ax2.plot(context_sizes, valid_completions_ar, color="green", label="AR")
ax2.plot(context_sizes, valid_completions, color="green", linestyle="dotted", label="Infilling")
ax2.set_xlabel("Context Size, tokens")
ax2.set_ylabel("Valid Completions, %")
ax2.set_title("Valid Completions")
ax2.legend()

# Set the x-ticks and labels
ax2.set_xticks(context_sizes)
ax2.set_xticklabels(context_sizes)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

solutions = ["Copilot", "Optimal", "Best", "Code Llama 13B"]
correct_completions = [29, 36, 41, 34]
valid_completions = [68, 61, 64, 64]
speed = [19, 81, 67, 68]

# Set the width of each bar
bar_width = 0.25

# Calculate the x-coordinates for the bars
x = np.arange(len(solutions))

# Create a figure and axis objects
fig, ax = plt.subplots(dpi=150)

# Plot the correct completions, valid completions, and speed as bars
ax.bar(x - bar_width, correct_completions, width=bar_width, color="blue", label="Correct Completions")
ax.bar(x, valid_completions, width=bar_width, color="green", label="Valid Completions")

# Set the x-ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(solutions)

# Set the y-axis label for the completions
ax.set_ylabel("Completions, %")

# Create a twin axis for the speed
ax2 = ax.twinx()
ax2.bar(x + bar_width, speed, width=bar_width, color="orange", label="Speed")
ax2.set_ylabel("Speed, char/s")

# Adjust the y-axis limits
ax.set_ylim([0, max(valid_completions) + 10])
ax2.set_ylim([0, max(speed) + 10])

# Combine the legends
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc="upper right")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
